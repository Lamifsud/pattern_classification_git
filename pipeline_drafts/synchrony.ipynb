```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
spk_number = np.arange(5)
print(spk_number)
```

```python
spk_number = np.arange(5)
print(spk_number)
spk_train1 = np.repeat(spk_number, len(spk_number)) # repeat each value
spk_train2 = np.tile(spk_number, len(spk_number)) # repeat the entire sequence



spk_rate = np.array((spk_train1, spk_train2))
n_times = spk_rate.shape[1]
sampling_rate = 4

time_points = np.linspace(0, n_times, num=n_times * sampling_rate)

spike_train = np.zeros((2, n_times * sampling_rate))
```

```python
time_points.shape
```

```python
# Fill in spike_train based on spike patterns
n_units = 2 
len_time = np.arange(0, time_points.shape[0], sampling_rate)


for i in range(n_units):
    print('unit', i)
    for j in range(len_time.shape[0]):
        if spk_rate[i][j] != 0:
            spk_count = spk_rate[i][j]
            spike_train[i, len_time[j]:len_time[j]+spk_count] = 1
```
